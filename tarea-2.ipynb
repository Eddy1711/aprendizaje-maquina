## Ejercicio de Simulación

Este ejercicio demuestra la validez de la norma vectorial $$\|x\|_\Sigma = \sqrt{x^T \Sigma x}$$, donde $$\Sigma$$ es una matriz simétrica y semidefinida positiva. Se genera una gráfica de contornos de nivel para ilustrar cómo se comporta esta norma en el plano $$xy$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir una matriz simétrica y semidefinida positiva
Sigma = np.array([[2, 0.5], [0.5, 1]])

# Definir una función para calcular la norma ||x||_Sigma
def norm_Sigma(x, Sigma):
    return np.sqrt(np.dot(np.dot(x.T, Sigma), x))

# Generar puntos en una rejilla
x_vals = np.linspace(-2, 2, 400)
y_vals = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Calcular la norma ||x||_Sigma para cada punto en la rejilla
Z = np.array([[norm_Sigma(np.array([x, y]), Sigma) for x in x_vals] for y in y_vals])
```

```python
# Graficar los contornos de nivel de la norma ||x||_Sigma
plt.contour(X, Y, Z, levels=20)
plt.title('Contornos de Nivel de la Norma ||x||_Sigma')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='||x||_Sigma')
plt.grid(True)
plt.show()
```
